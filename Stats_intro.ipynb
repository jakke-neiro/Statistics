\subsection{Interval estimators of a population proportion}
We assume $n$ independent trials, each with (unknown) probability of success $p$. Observe $X$ successes. The confidence interval for the proportion is
\[
\hat{p} \pm z_{\alpha/2} \sqrt{ \frac{\hat{p}(1-\hat{p})}{n} }
\]

\begin{minted}[breaklines]{python}
p_hat = 69/339
[p_hat - scs.norm.ppf(0.975)*np.sqrt((p_hat*(1-p_hat))/339), p_hat + scs.norm.ppf(0.975)*np.sqrt((p_hat*(1-p_hat))/339)])

from statsmodels.stats.proportion import proportion_confint
proportion_confint(69, 339)
\end{minted}

\section{Hypothesis tests concerning two populations}
\subsection{Paired-sample t test}
$X_1, \dots, X_n$ and $Y_1, \dots, Y_n$, $H_0: \mu_X = \mu_Y$ and $H_1: \mu_X \neq \mu_Y$, $D_i = X_i - Y_i$.

Reject $H_0$ if
\[
\sqrt{n}\frac{\overline{D}}{S_d} \geq t_{n_1, \alpha/2}
\]

\begin{minted}[breaklines]{python}
before = np.array([11.2, 9.4, 9.9, 9.3, 8.9, 8.2, 10.5, 8.8, 10.3, 9.8])
after = np.array([12.9, 10.8, 10.3, 10.9, 8.5, 8.9, 10.4, 8.5, 11.2, 10.1])

mean_diff = np.mean(before-after)
s_d = np.std(before-after, ddof=1)
stat = np.sqrt(len(before))*(mean_diff/s_d)
print("Paired test statistic" + str(stat))
print("P-value: " + str(2*(1-scs.t.cdf(np.abs(stat), len(before)-1))))

print(scs.ttest_rel(before, after))
\end{minted}

\section{Chi-squared goodness-of-fit tests}
\subsection{Chi-square in python}
\subsubsection{Goodness of fit}
The default degrees of freedom, k-1, are for the case when no parameters of the distribution are estimated. If p parameters are estimated by efficient maximum likelihood then the correct degrees of freedom are \textbf{k-1-p}! If the parameters are estimated in a different way, then the dof can be between k-1-p and k-1. However, it is also possible that the asymptotic distribution is not a chisquare, in which case this test is not appropriate.

\begin{minted}[breaklines]{python}
observed = np.array([5267, 4436, 1800, 534, 111, 21])
expected = np.array([5253.48535961, 4412.92770207, 1853.42963487,  518.96029776, 108.98166253, 18.30891931])

chi_stat = np.sum((observed-expected)**2/(expected))
1-scs.chi2.cdf(chi_stat, 4) #p-value

import scipy.stats as scs
scs.chisquare(observed, expected, ddof=1)
\end{minted}
