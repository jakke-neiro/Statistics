\subsection{Matemaattisen todennäköisyyden käsite}
P on todennäköisyys, mikäli se toteuttaa seuraavat kolme ehtoa

\begin{enumerate}
    \item $P(A) \geq 0$ kaikille tapahtumille $A \subset \Omega$
    \item $P(\emptyset) = 0$ ja $P(\Omega) = 1$
    \item Täysadditiivisuus $P\bigg( \bigcup_{j=1}^{\infty}A_j \bigg)$, kun $A_1, A_2, \dots$ ovat erilliset
\end{enumerate}

Todennäköisyyden ominaisuuksia

\begin{enumerate}
    \item Äärellinen additiivisuus: Jos $A \cap B = \emptyset$, niin $P(A \cup B) = P(A) + P(B)$
    \item $P(A^{\mathsf{c}} ) = 1 - P(A)$ kaikille $A$
    \item $0 \leq P(A) \leq 1$ kaikille A
    \item Jos $A \subset B$, niin $P(A) \leq P(B)$
    \item $P(A\setminus B) = P(A) - P(A \cap B)$ kaikille $A$ ja $B$
    \item Yhteenlaskukaava: $P(A \cup B) = P(A) + P(B) - P(A \cap B )$ kaikille $A$ ja $B$
    \item Boolen epäyhtälö: $P(A \cup B) \leq P(A) + P(B)$
    \item Bonferronin epäyhtälö: $P(A \cap B) \geq P(A) + +P(B) -1$
\end{enumerate}

$\sigma$-algebra: todennäköisyys on määritelty kaikille meitä kiinnostaville perusjoukon osajoukoille

\section{Satunnaismuuttuja}
\subsection{Satunnaismuuttuja ja sen jakauma}
$\Omega$ on perusjoukko. $X: \Omega \to \mathcal{R}$ on satunnaismuuttuja ($X$ on Borelin funktio). 

Tapahtuman indikaattori $I_A(\omega) = \begin{cases}
1 & \textrm{jos} \omega \in A \\
0 & \textrm{muuten}
\end{cases}$

Millä todennäköisyydellä $X$ saa arvon joukosta $B$, $B \subset \mathcal{R}$?

$\{\omega \in \Omega: X(\omega) \in B \} = \{X \in B \}$

$P(X \in B) = P(\{\omega \in \Omega: X(\omega) \in B\})$

$P(X=X) = P(\{\omega \in \Omega: a \leq X(\omega) \leq b\})$

$P(a \leq X \leq b) = P(\{\omega \in \Omega: a \leq X(\omega) \leq b \})$

$P(X \leq x) = P(\{\omega \in \Omega: X(\omega) \leq x \})$

Jakauma: $B \subset \mathcal{R} \to P(X \in B)$, Borelin funktio, Perusjoukon osajoukot ovat kaikki tapahtumia

\subsection{Kertymäfunktio}
\textbf{Cumulative distribution function, Kertymäfunktio, Fördelningsfunktion}

Kertymäfunktio: $F: \mathcal{R} \to \mathcal{R}$, $F(x) = P(X \leq x)$, $x \in \mathcal{R}$

\begin{enumerate}
    \item $F$ on kasvava funktio
    \item $F$ on oikealta jatkuva funktio eli $F(x+)=F(x)$, $\forall x \in \mathcal{R}$
    \item $F(-\infty) = 0$ ja $F(\infty) = 1$
\end{enumerate}

Kääntäen, jos funktiolla $F: \mathcal{R} \to \mathcal{R}$ on nämä ominaisuudet, niin on olemassa sm $X$ siten, että $F$ on $X$:n kertymäfuntkio.

Sattumistodennäköisyydet väleille: $P(a \leq X \leq b) = F(b) - F(a)$, koska $\{X \leq b \} = \{\{X \leq a \} \cup \{a \leq X \leq b\}$, jossa osat ovat erilliset

Jos $F$ on satunnaismuuttujan $X$ kertymäfunktio, niin $\forall x$ $P(x < x) = F(x-)$

$P(X = x) = P(X \leq x) - P(X < x) = F(x) - F(x-)$

Merkintä: $F_x$ on $X$:n kertymäfunktio ja $F_Y$ on $Y$:n kertymäfunktio

Sama jakauma, jos $P(X \in B) = P(Y \in B)$, $\forall B \subset \mathcal{R}$, $X = Y \to F_X = F_Y$

\subsection{Diskreetti jakauma}
Satunnaismuuttujan $X$ jakauma on diskreetti, jos sen arvojoukko $\{x_1, x_2, \dots \}$ on äärellinen tai korkeintaan numeroituvasti ääretön.

Satunnaismuuttujan $X$ pistetodennäköisyysfunktio (probability mass function, sannolikhetsfunktion) on funktio $f: \mathcal{R} \to \mathcal{R}$, jolle $f(x) = P(X=x)$, $x \in \mathcal{R}$

Merkintä: muilla $x \in \mathcal{R}$ on $P(X=x)=0$, $f_x$

$0 \leq f(x) \leq 1$, $\forall x$ ja $f(x)=0$, kun $x \notin S$

$f$ määrää $X$:n jakauman kaavalla $P(X \in B) = \sum_{x \in B}f(x)$

Olkoon $f: \mathcal{R} \to \mathcal{R}$ ei-negatiivinen funktio, joka saa positiivisia arvoja korkeintaan numeroitumassa $\mathcal{R}$:n osajoukossa siten, että $\sum_xf(x)=1$, tällöin se diskreetin satunnaismuuttujan pistetodennäköisyysfunktio. 

\subsection{Esimerkkejä diskreeteistä jakaumista}
Diskreetti tasajakuma, Bernoullin jakauma, Binomijakauma, Hypergeometrinen jakauma, Negatiivinen binomijakauma, Geometrinen jakauma, Poissonin jakauma

Diskreetti tasajakuma: satunnaismuuttuja $X$ joukossa $\{1, \dots, N\}$

$f(x) = \frac{1}{N}$, kun $x=1,2,\dots, N$

Bernoullin jakauma: Olkoon $A$ tapahtuma, ja merkitään $p=P(A)$, jolloin $0 \leq p \leq 1$. Indikaattorilla $X = I_A$ on Bernoullin jakauma parametrilla $p$, $X \sim \textrm{Bernoulli}(p)$. 

Huom! $\textrm{Bernoulli}(p) = \textrm{Bin}(1, p)$

\subsection{Integraalilaskennan kertausta}
Riemannin integraali tuttu, stokastiikassa käytetään Lebesguen integraalia.

Antiderivaatta eli integraalifunktio: Olkoon $f$ välillä $I$ määritelty reaalifunktio: $F'(x) = f(x)$, $\forall x \in I$. 

Määrätty integraali
\[
\int_a^bf(x)dx = \Big|_a^b F(x) = F(b) - F(a)
\]

Osittaisintegrointi:
\[
\int_a^b u(x)v'(x) dx = \Big|_a^b u(x)v(x) - \int_a^b u'(x)v(x)dx

Sijoituskeino:

\int_a^b f(x) dx = \int_c^d f(g(t))g'(t)dt

\int_0^{\pi/2}(\sin2x)dx = \frac{1}{2}\int_{x = 0}^{x = \pi/2} \sin2x 2 dx = \frac{1}{2} \int_{u=0}^{u = \pi} \sin u du = \frac{1}{2}(-\cos\pi + \cos=) = 1

\subsection{Jatkuva jakauma}
Satunnaismuuttujalla $X$ on jatkuva jakauma tiheysfunktiolla $f$, jos $f$ on ei-negatiivinen ja jos $\forall a < b$
\[
P(a \leq X \leq b) = \int_a^bf(x)dx
\]

Integraalikäsite: Lebesguen integraali, pilkotaan Riemannin integraaliksi

Tiheysfunktion yksikäsitteisyys: ei-negatiiviset funktiot $f$ ja $g$ ovat saman jakauman tiheysfunkioota. jos ja vain jos ne yhtyvät melkein kaikkialla, eli joukko jossa $f \neq g$ on nollamittainen 

Välien sattumistodennäköisyydet:
\[
P(X = x) = P(x \leq X \leq x) = \int_x^xf(u)du = 0
\]

Kertymäfunktion ja tiheysfunktion yhteys:

Jos $X$:n tiehysfunktio on $f$, niin sen kertymäfunktio $F$ on $F(x) = \int_{-\infty}^{\infty}f(u)du$

Jos $X$:n kertymäfunktio on $F$, niin $F$ on derivoituva melkein kaikkialla, ja $X$:n tiheysfunktioksi voidaan valita $F'$.

Jatkuvan jakauman tunnistaminen kertymäfunktiosta:
\begin{enumerate}
    \item $F$ on jatkuva koko reaaliakselilla
    \item $F$ on derivoituva kaikkialla muualla paitsi mahdollisesti äärellisessä määrässä pisteitä
    \item $f = F'$ on jatkuva kaikkialla muualla paitsi mahdolllisesti äärellisessä määrässä pisteitä
\end{enumerate}

Absoluuttisesti jatkuva jakauma

Tiheysfunktion tunnistus: $\int_{-\infty}^{\infty}f(x)dx = P(\Omega) = 1$

Normalisoimaton tiheysfunktio: Jos $g \geq 0$ on sellainen reaaliakselilla määritelty funktio, jonka integraali $\int_{-\infty}^{\infty}g(x)dx$ on äärellinen ja aidosti positiivinen, niin tällöin on olemassa vakio $k$ siten, että $f = kg$ on tiheysfunktio. 
\[
\frac{1}{k} = \int_{-\infty}^{\infty}g(x)dx, \int f(x) dx = k \int g(x) dx = \frac{k}{k} = 1
\]
Olkoon satunnaismuuttujalla $X$ tiheysfunktio $f$, jota emme tunne. Oletetaan, että meillä on aineisto $x_1, x_2, \dots x_:n$ jota voidaa pitää $N$ riippumattoman satunnaismuuttujan $X$ jakaumaa noudattavan satunnaismuuttujan havaittuina arvoina.
\[
x_1 = X_1(\omega^{act}), x_2 = X_2(\omega^{act}), \dots x_N = X_n(\omega^{act})
\]

Mikä on tiheysfunktion arvo pisteessä $u$?

Valitaan pieni luku $h > 0$, $[u-h/2, u+h/2]$, lukumäärä $N_{u, h}$

Frekvenssitulkinta: $P(u - h/2 \leq X \leq u + h/2) = \int_{u-h/2}^{u + h/2} f(x) dx \approx \frac{N_{u, h}}{N}$

Väliarvolause: $\int_{u - h/2}^{u + h/2}f(x)dx \approx hf(u)$, $f(u) = \frac{N_{u, h}}{hN}$

\subsection{Satunnaismuuttujan muunnos}
Jos $X$ on satunnaismuuttuja ja $g: \mathcal{R} \to \mathcal{R}$ on funktio, niin myös $Y = g(x)$ on satunnaismuuttuja. 

$Y = g(X)$, $Y(\omega) = g(X(\omega))$, $\forall \omega \in \Omega$

Alkukuva: $g: \mathcal{R} \to \mathcal{R}$ on funktio ja $A \in \mathcal{R}$, niin merkintä $g^{-1}(A)$ tarkoittaa joukon $A$ alkukuvaa kuvauksessa $g$, eli $g^{-1}(A) = \{x \in \mathcal{R}: g(X) \in A \}$ ja $g(X) \in A \iff x \in g^{-1}(A)$. 

Muunnoksen $g(X)$ jakauma diskreetissä tapauksessa:

Olkoon $X$ diskreetti satunnaismuuttuja pistetodennäköisyysfunktiolla $f_X$, ja olkoon $g: \mathcal{R} \to \mathcal{R}$ funktio. Tällöin satunnaismuuttuja $Y = g(X)$ on diskreetti, ja sen pistetodennäköisyysfunktio $f_Y$ on
\[
f_Y(y) = \sum_{x \in g^{-1}(\{ y \})} f_X(x)
\]

\begin{proof}
\[
P(Y=y) = P(g(X) = y) = P(g(X) \in \{y\}) = P(X \in g^{-1}(\{y\}))
\]
\end{proof}

Erikoistapaus: Jos yhtälöllä $y = g(X)$ on yksikäsitteinen ratKAISU $x = h(y)$ kaikilla $y$, jotka kuuluvat satunnaismuuttujan $Y$ arvojoukkoon $S_Y$, niin
\[
f_Y(y) = f_X(h(y)), y \in S_Y
\]

Jatkuvasti jakautuneen satunnaismuuttujan muunnos $\to$ diskreetti, jatkuva tai sekatyyppinen 

\subsection{Kvantiilifunktio ja sen käyttö simuloinnissa}
\textbf{Inverse CDF or quantile function, Kvantiilifunktio, Kvantilfunktion}

Oletusten ansiosta funktiolla $F$ on olemassa käänteisfunktoi $F^{-1}(0,1) \to (a, b)$. Merkintä $F^{-1}$ ei tarkoita funktiota $1/F$, vaan kertymäfunktion käänteisfunktiota!
\[
F^{-1}(u) \iff F(x) = u, 0 < u < 1
\]

Kertymäfunktion käänteisfunktio = kvantiiliifunktio

Osuus $u$ jakauman todennäköisyysmassasta jää pisteen $F^{-1}(u)$ vasemmalle puolelle eli vasemmanpuoleiseen häntään: jos satunnaismuuttujalla $X$ on kertymäfunktio $F$, niin
\[
P(X \leq F^{-1}(u)) = F(F^{-1}(u)) = u
\]
\[
u = P(X \leq F^{-1}(u)) = \int_{-\infty}^{F^{-1}(u)}f(x)dx
\]
Jos $0 < u <1$, niin jakauman $u$-alakvantiili, josta vasemmalle jää jakaumansta osuus $u$: $F^{-1}(u)$

Jakauman $u$-yläkvantiili taas on sellainen piste, josta oikealle jää jakaumasta osuus $u$, joten $F^{-1}(1-u)$.

Kvantiili: mediaani, alakvartiili, yläkvartiili, kvintiili, desiili ja persentiili

Kertymäfunktion ominaisuuksia: Funktio $F^{-1}$ on kasvava funktio!
\[
F^{-1}(F(x)) = x, \forall a < x <b
\]
\[
F(F^{-1}(u)) = u \forall 0 < u <1
\]

Kertymäfunktion muunnos: Olkoon $X$ satunnaismuuttuja, jolla on kertymäfunktio $F$. Tarkastellaan satunnaismuuttujaa $F(x)$.

\[
P(F(X) \leq u) = u
\]

\[
F(X) \sim U(0, 1)
\]

Käänteisfunktiomenetelmä, satunnaismuuttuja $Z = F^{-1}(u)$. 

\begin{example}
\[
F(x) = \begin{cases}
1 - e^{-x} & \textrm{ jos } x \geq 0 \\
0 & \textrm{muuten}
\end{cases}
\]

$F(x) = u \iff X=+\ln(1-u)$

$\textrm{Exp}(1)$ voidaan simuloida asettamalla $X = -\ln(1-u)$, kun $u$ on ensin generoitu jakaumasta $U(0,1)$. 
\end{example}

\section{Yhteisjakauma}
\textbf{Joint distribution, yhteisjakauma, simultanfördelning}
\subsection{Kaksiulotteinen satunnaisvektori ja sen jakauma}
\textbf{Bivariate distribution, Kaksiulotteinen jakauma, Tvådimensionell fördelning}

In the continuous case, we call a function $f(x, y)$ a PDF for the random variables $(X, Y)$ if 1. $f(x, y) \geq 0$ for all $(x, y)$, 2. $\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f(x, y)dxdy = 1$ and 3. for any set $A \subset \R \times \R, P((X, Y) \in A) = \iint_A f(x, y) dxdy$ 

Olkoot $X$ ja $Y$ samalla perusjoukolla $\Omega$ määriteltyjä satunnaismuuttujia, $X: \Omega \to \mathcal{R}$ ja $Y: \Omega \to \mathcal{R}$. Tällöin $(X, Y)$ on kaksiulotteinen satunnaisvektori $(X, Y): \Omega \to \mathcal{R}^2$, $(X, Y)(\omega) = (X(\omega), Y(\omega))$.

Satunnaisvektorin $(X, Y)$ jakauma eli yhteisjakauma on $\mathcal{R}^2$:n osajoukoilla $A$ määritelty funktio
\[
P((X, Y) \in A) = P(\{ \omega \in \Omega : (X(\omega), Y(\omega) ) \in A \} )
\]

\textbf{Joint distribution, Yhteisjakauma, Simultanfördelning}

Merkintäsopimus: $P(X \in A, Y \in B) = P(\{X \in A \} \cap \{Y \in B \})$

Kertymäfunktio: $F_{X, Y}(x, y) = F(x, y) = P(X \leq x, Y \leq y)$, $x, y \in \mathcal{R}$.

Yhteiskertymäfunktio määrää jakauman.

Yksittäisen satunnaismuuttujan jakauma = reunajakauma

\textbf{Marginal distribution, Reunajakauma, Marginalfördelning}

$F_X(x) = P(X \leq x, Y \in \mathcal{R}) = \lim_{y \in \infty}F_{X, Y}(x, y) = F_{X, Y}(x, \infty)$

$F_Y(y) = \lim_{x \in \infty}F_{X, Y}(x, y) = F_{X, Y}(\infty, y)$

If $(X, Y)$ have joint distribution with mass function $f_{X, Y}$, then the marginal mass function for $X$ is defined by
\[
f_X(x) = P(X = x) = \sum_y P(X=x, Y=y) = \sum_y f(x, y)
\]
and the marginal mass function for $Y$ is defined by
\[
f_Y(y) = P(Y = y) = \sum_x P(X=x, Y=y) = \sum_x f(x, y)
\]
For continuous random variables, the marginal densities are
\[
f_X(x) = \int f(x, y) dy
\]
and
\[
f_Y(y) = \int f(x, y)dx
\]
the corresponding marginal distribution functions are denoted by $F_X$ and $F_Y$.

The expected value, or mean, or first moment, of $X$ is defined to be
\[

\int x f(x) dx \\

\[

Suurten lukujen laki

Binomijakauman odotusarvo: $X \sim \textrm{Bin}(n, p)$, $q = 1 - p$

\[

\textbf{Markov's inequality, Markovin epäyhtälö, Markovs olikhet}

\end{proof}

\section{Kaksiulotteinen jakauma}
\subsection{Jatkuva kaksiulotteinen jakauma}
Määritelmä: Jatkuva (yhteis)jakauma, yhteistiheysfunktio

Satunnaisvektorilla $(X, Y)$ on jatkuva jakauma, eli satunnaismuuttujilla $X$ ja $Y$ on jatkuva yhteisjakauma, mikäli on olemassa $\mathcal{R}^2$:lla määritelty funktio $f \geq 0$
\[
P((X, Y) \in B) = \iint_B f(x, y) dx dy, B \subset \mathcal{R}^2
\]
$f = f_{X, Y}$ on tiheysfunktio, yhteistiheysfunktio

Tasointegraali: $\int_B g = \int_B g(x, y) d(x, y) = \iint_B g(x, y) dxdy$

Fubinin lause, iteroitu integraali, Lebesguen integraalit

Fubini: Seuraavissa tapauksissa tasointegraali voidaan laskea iteroituvana integraalina, eli alla olevat yhtälöt ovat voimassa
\[
\int_B g(x, y) d(x, y) = \int_{\infty}^{\infty} \bigg( \int_{-\infty}^{\infty} I_B(x, y)g(x, y)dx \bigg) dy = \int_{\infty}^{\infty} \bigg( \int_{-\infty}^{\infty} I_B(x, y)g(x, y)dy \bigg) dx
\]

Jos $g \geq 0$, jolloin integraalien yhteinen arvo voi olla $\infty$

Jos $\int_B|g| < \infty$, jolloin integraalien yhteinen arvo on äärellinen. 

\[
\iint_B g(x, y) dx dy = \int_{a}^b \bigg( \int_{c(x)}^{d(x)}g(x, y)dy \bigg) dx = \int_{c}^d \bigg( \int_{a(y)}^{b(y)}g(x, y)dx \bigg)dy
\]

Olkoon satunnaisvektorilla $(X, Y)$ jatkuva jakauma. Tällöin reunajakaumat ovt myös jatkuvia, ja niiden tiheysfunktiot saadaan integroimalla toinen muuttuja pois yhteistiheysfunktiosta
\[
f_X(x) = \int_{-\infty}^{\infty}f_{X, Y}(x, y)dy
\]
\[
f_Y(y) = \int_{-\infty}^{\infty}f_{X, Y}(x, y)dx
\]

Kertymäfunktio:
\[
F_{X, Y}(x, y) = P((X, Y) \in (-\infty, x] \times (-\infty, y]) = \int_{-\infty}^{\infty}ds\int_{-\infty}^{\infty}f_{X, Y}(s, t)dt
\]
\[
\frac{\partial}{\partial x}F_{X, Y}(x, y) = \int_{-\infty}^y f_{X, Y}(x, t)dt
\]
\[
\frac{\partial^2}{\partial x \partial y} F_{X, Y}(x, y) = f_{X, Y}(x, y)
\]

Jos satunnaisvektorilla $(X, Y)$ on jatkuva jakauma, niin jakauman tiheysfunktioksi voidaan valita $f_{X, Y}(x, y) = \frac{\delta^2}{\delta_x \delta_y} = F_{X, Y}(x, y)$

Jos $f: \R^2 \to \R$ on ei-negatiivinen, ja sen integraali koko tason yli on yksi, niin se on jonkin kaksiulotteisen satunnaisvektorin tiheysfunktio.

\subsection{Ehdolliset jakaumat}
Jos satunnaisvektorilla $(X, Y)$ on diskreetti jakauma, niin satunnaismuuttujan $X$ ehdollinen pistetodennäköisyysfunktio ehdolla $Y = y$ on
\[
f_{X|Y} = (x|y) = P(X=x | Y=y) = \frac{P(X = x, Y = y)}{P(Y = y)} = \frac{f_{X, Y}(x, y)}{f_Y(y)}
\]

Määritelmä: Olkoon $(X, Y)$:llä jatkuva jakauma tiheysfunktiolla $f_{X, Y}$. Tällöin satunnaismuuttujan $X$ ehdollinen tiheysfunktio ehdolla $Y=y$ on 
\[
f_{X|Y}(x|y) = \frac{f_{X, Y}(x, y)}{f_Y(y)}
\]
$x \in \mathcal{R}$ kun $f_Y(y) > 0$.
\[
f_{Y|X}(y|x) = \frac{f_{X, Y}(x, y)}{f_X(x)}
\]
$y \in \mathcal{R}$ ja $f_X(x) > 0$.

$P(X \in A| y \leq Y \leq y + h) \to \int_A f_{X|Y}(x|y)dx$

\subsection{Kertolaskusääntö eli ketjusääntö}
$f_{X, Y}(x, y) = f_X(x)f_{Y|X}(y|x) = f_Y(y) = f_{X|Y}(x|y)$, $\forall x, y$

Rajaudutaan tapauksiin 
\[
f_{Y|X}(y|x) = \begin{cases}
\frac{f_{X, Y}(x, y)}{f_X(x)} & f_X(x) > 0 \\
0 & \textrm{muuten}
\end{cases}
\]

\[
f_{Y|X}(x) = \frac{f_{X, Y}(x, y)}{f_X(x)} = \frac{f_Y(y)f_{X|Y}(x|y)}{f_X(x)} \propto f_Y(y)f_{X|Y}(x|y) = f_{X, Y}(x, y)
\]

\subsection{Diskreetin ja jatkuvan muuttujan yhteisjakauma}
$X$ on diskreetti, $Y$ on jatkuva. $X$: n arvot $x_1, x_2, \dots$

$f_{Y|X}(y|x_i), i \geq 1$, $P(X = x_i, Y \in B) = f_X(x_i)\int_B f_{Y|X}(y|x_i)dy$

$f_{X, Y}(x, y) = f_X(x)f_{Y|X}(y|x)$, $P(X \in A, Y \in B) = \sum_{x \in A}\int_B f_{X, Y}(x, y)dy$, $A, B \subset \mathcal{R}$

Olkoon $(X, Y)$ satunnaisvektori, jossa $X$:llä on diskreetti ja $Y$:llä jatkuva jakauma, ja olkoon $Z = g(X, Y)$ jokin sen reaaliarvoinen muunnos

\]
mikäli $\sum_x \int|g(x, y)|f_{X, Y}(x, y)dy < \infty$

\subsection{Ehdollinen odotusarvo}

\]
\[

\]
\[
\textrm{var}(g(X, Y) | X = x) = E(g(X, Y)^2 | X = x) - \bigg(E(g(X, Y) |X = x) \bigg)^2 
\]

$X \to E(Y|X = X)$ on regressiofunktio

$E(Y|X = x) ) = \int y f_{Y|X}(y|x)dy$, $E(Y|X=x) = \sum_y y f_{Y|X}(y|x)$

Määritelmä: Ehdollinen odotusarvo ehdolla satunnaismuuttuja

$m(x) = E(g(X, Y)|X=x)$, satunnaismuuttuja $E(Y | X=x) = \sum_y y f_{Y|X}(y|x)$

\section{Raja-arvolauseita ja approksimaatioita}
The most important aspect of probability theory concerns the behavior of sequences of random variables. This part of probability is called large sample theory, or limit theory, or asymptotic theory. 

The law of large numbers says that the sample average $\overline{X}_n = \frac{1}{n}\sum_{i=1}^n X_i$ converges in probability to the expectation $\mu = E(X_i)$. This means that $\overline{X}_n$ is close to $\mu$ with high probability. 

The central limit theorem says that $\sqrt{n}(\overline{X}_n - \mu)$ converges in distribution to a normal distribution. This means that the sample average has approximately a normal distribution for large $n$.  

\subsection{Suurten lukujen laki}
\textbf{Law of large numbers, Suurten lukujen laki, Lagen om stora tal}

\textbf{Convergene in probability, stokastinen suppeniminen, Stokastisk konvergens}

Jono satunnaismuuttujia $X_1, X_2, \dots$ suppenee stokastisesti eli konvergoi stokastisesti kohti satunnaismuuttujaa $Y$, jos
\[
P(|X_n - Y| \geq \epsilon) \to 0, n \to \infty, \forall \epsilon > 0
\]
$X_n \xrightarrow{p} Y$ tai $\lim_{n \to \infty}X_n = Y$

Let $X_1, X_2, \dots$ be a sequence of random variables and let $X$ be another random variable. Let $F_n$ denote the CDF of $X_n$ and let $F$ denote CDF of $X$. 

$X_n$ converges to $X$ in probability, written $X_n \xrightarrow{p} X$, if for every $\epsilon > 0$ 
\[
P(|X_n - X| > \epsilon) \to 0
\]
as $n \to \infty$. 

\textbf{Almost sure convergence, Melkein varma suppeneminen, Nästan säker konvergens}

Jono satunnaismuuttujia $X_1, X_2, \dots$ suppenee melkein varmasti kohti satunnaismuuttujaa $Y$, jos $X_n(\omega) \to Y(\omega)$ perusjoukon osajoukossa, jonka todennäköisyys on yksi, eli jos
\[
P(\lim_{n \to \infty} X_n = Y) = 1
\]

\textbf{Strong law of large numbers, Vahva suurten lukujen laki, De stora talens lag i stark form}

Olkoon $X_1, X_2, \dots$ jono riippumattomia ja samoin jakautuneita satunnaismuuttujia, joiden odotusarvo $\mu = EX$ on olemassa. Tällöin keskiarvojen $\overline{X}_n = \frac{1}{n}\sum_{i=1}^n X_i$ muodostama jono suppenee melkein varmasti kohti arvoa $\mu$, eli $\overline{X}_n \xrightarrow{p} \mu$ 

\subsection{Jakaumasuppeneminen}
\textbf{Convergence in distribution, Jakaumasuppeneminen, Konvergens mot fördelning}

Olkoon $X_1, X_2, \dots$ jono satunnaismuuttujia, joiden kertymäfunktiot ovat $F_1, F_2, \dots$. Olkoon $Y$ satunnaismuuttuja, jonka kertymäfunktio on $G$, ts. kaikilla $x$ $F_n(x) = P(X_n \leq x), \sigma(x) = P(Y \leq x)$.

Jono $(X_n)$ suppenee jakaumaltaan kohti $Y$:tä, mikäli $\lim_{n \to \infty}F_n(x) = G(x)$ kaikissa rajajakauman kertymäfunktion $G$ pisteissä $x$. 
\[
X_n \to N(0, 1), X_n \approx Y
\]
Joudutaan tarkkojen luottamusvälien sijasta soveltamaan tähän ajatukseen perustuvia asymptoottisia luottamusvälejä.

$X_n$ converges to $X$ in distribution, written $X_n \xrightarrow{d} X$ , if 
\[
\lim_{n \to \infty}F_n(t) = F(t)
\]
at all $t$ for which $F$ is continuous.
