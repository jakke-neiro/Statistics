\textbf{Sample space, Perusjoukko, Utfallsrum} $\Omega$: the set of possible outcomes of an experiment

\textbf{Events, Tapahtumat, Händelser}: subsets of $\Omega$

We will assign a real number $P(A)$ to every event $A$, called the probability of $A$. We also call $P$ a probability distribution or a probability measure. To qualify as a probability, $P$ must satisfy three axioms.

A function $P$ that assigns a real number $P(A)$ to each event $A$ is a probability distribution or a probability measure if it satisfies the following three axioms:
\begin{enumerate}
    \item $P(A) \geq 0$ for every $A$
    \item $P(\Omega) = 1$
    \item If $A_1, A_2, \dots$ are disjoint then
    \[
    P(\bigcup_{i=1}^{\infty}A_i) = \sum_{i=1}^{\infty}P(A_i)
    \]
\end{enumerate}
The two common interpretations are frequencies and degrees of beliefs. In statistical inference, the differing interpretations lead to two schools of inference: the frequentist and the Bayesian schools. 

Continuity of probabilities: If $A_n \to A$, then
\[
P(A_n) \to P(A)
\]
as $n \to \infty$.
