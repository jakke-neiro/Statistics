\section{Tilastollinen malli}
\subsection{Tilastollinen malli}
\textbf{Statistical model, Tilastollinen malli, Statistisk model}

Numeerinen aineisto $\overline{y} = (y_1, \dots, y_n)$, havainnot $y_i$

Satunnaismuuttujien $Y_1, \dots, Y_n$ realisaatiot $y_1, \dots, y_n$, $Y(\omega_{act} = y$

Tilastollinen päättely: aineiston $y$ perusteella päätellä todennäköisyysjakauma, jota $\overline{Y} = (Y_1, \dots, Y_n)$ noudattaa

Yhteispistetodennäköisyysfunktio, yhteistiheysfunktio

Satunnaismuuttujat $Y_1, \dots, Y_n$ ovat riippumattomat
\[
f_{\overline{Y}}(\overline{y}) = f_{Y_1}(y_1) \dots f_{Y_n}(y_n)
\]
Parametrinen päättely: oletetaan, että tilastollinen malli on tunnettu yhtä tai useampaa reaalista parametria vaille

$f_{\overline{Y}}(\overline{y}; \theta)$, $\overline{theta} = (\theta_1, \dots, \theta_d)$ parametrinen malli

$\Omega \subset \mathcal{R}^d$ parametriavaruus, parametrinen päättely

\subsection{Esimerkkejä parametrisista malleista}
\subsubsection{Toistokoemalli}
$n$ otos, $y_i = \begin{cases} 0 & \textrm{ehjä} \\ 1 & \textrm{rikki} \end{cases}$, $y = (y_1, \dots, y_n)$

$f_{Y_i}(y_i; \theta) = \theta^{y_i}(1 - \theta)^{1 - y_i}$, $y_i = 0,1$ Bernoulli-jakauma

$Y_1,\dots, Y_n \independent$, $f_{\overline{Y}}(\overline{y}; \theta) = \prod_{i=1}^n f_{Y_i}(y_i; \theta) = \theta^k (1 - \theta)^{n - k}$, $k = y_1 + \dots + y_n$

\subsubsection{Kestoiät}
$Y_1, \dots, Y_n \sim \textrm{Exp}(\lambda) \independent$

$f_{\overline{Y}}(\overline{y}; \lambda) = \prod_{i = 1}^n \lambda e^{-\lambda y_i} = \lambda^n e^{-\lambda(y_1 + \dots + y_n)} = \lambda^n e^{-\lambda n \overline{y}}$

\subsubsection{Normaalihavainnot}
$Y_1, \dots, Y_n \sim N(\mu, \sigma^2) \independent$ parametriavaruus $\mathcal{R} \times (0, \infty)$

$f_Y(y; \mu, \sigma^2) = \prod_{i = 1}^n \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-(y_i - \mu)^2/\sigma^2} = \frac{1}{(2 \pi \sigma^2)^{n/2}} \exp \bigg( - \frac{1}{2 \sigma^2} \sum_{i=1}^n (y_i - \mu)^2 \bigg)$

\subsubsection{Lineaarinen regressiomalli}
$E(Y_i) = \alpha + \beta x_i$, $Y_i \sim N(\alpha + \beta x_i, \sigma^2)$, $Y_1, \dots, Y_n \independent$

$f_{\overline{Y}}(\overline{y}; \alpha, \beta, \sigma^2) = \frac{1}{(2 \pi \sigma^2)^{n/2}} \exp \bigg(- \frac{1}{2\sigma^2} \sum_{i = 1}^n (y_i - \alpha - \beta x_i)^2 \bigg)$

\subsection{Parametrisen päättelyn tavoitteet}
\begin{enumerate}
    \item Piste-estimointi: estimaatti tuntemattomalle parametrille $\theta$, estimaattien generointi ja hyvyyden mittaus
    \item Luottamusvälit ja -joukot, parametriavaruuden $\Omega$ osajoukko joka sisältää parametrin $\theta$ oikean arvon, väliestimointi
    \item Hypoteesin testaaminen: onko aineisto sopusoinnussa nollahypoteesin $\theta \in \Omega_{0}$ vai vastahypoteesin $\theta \in \Omega_1$ kanssa. $\Omega = \Omega_0 \cup Omega_1$ ja $\Omega_0 \cap \Omega_1 = \emptyset$
    \item Ennustaminen
    \item Mallin sopivuuden ja riittävyyden estimointi: $E(Y_i) = \alpha + \beta x_i$. Tilastollinen päättely edellyttää mallin spesifiointia, mallin arviointi on mahdollista analyysin jälkeen
\end{enumerate}

Tilastollinen malli $f_Y(y; \theta)$, suurten lukujen laki

Aineisto kerätään uudelleen ja uudelleen

Bayesiläinen päättely: parametri $\theta$ on satunnaisvektorin arvo

Priorijakauma $f_{\theta}$, $f_{\overline{Y}}(\overline{y}; \theta) = f_{Y|\Theta}(y | \theta)$

\[
f_{\Theta | Y}(\theta |y) =\frac{f_{\Theta}(\theta)f_{Y|\Theta}(y|\theta)}{f_{Y}(y)}
\]
\[

\]

\subsection{Pistemäärä ja säännölliset mallit, d=1}
Pistemääräfunktio: malli $f_{\overline{Y}}(\overline{y}; \theta)$, parametriavaruus $\Omega \subset \mathcal{R}$

Uskottavuusfunktio $\theta \to f_{\overline{Y}}(\overline{y}; \theta)$ on derivoituva $\Omega$:ssa $\forall y$

$l'(\theta) = l'(\theta; y) = \frac{\frac{\delta}{\delta \theta} f_{\overline{Y}}(\overline{y}; \theta)}{f_{\overline{Y}}(\overline{y}, \theta)}$, $\theta \in \Omega$ 

\textbf{Score, Pistemäärä, Skattningsfunktio}

Säännölliset mallit:

Jaktuva malli
\begin{enumerate}
    \item Jakauman kantaja (alusta) $A = \{y; f_{\overline{Y}}(\overline{y}, \theta) > 0\}$ ei riipu parametrista $\theta$
    \item $\theta \to f_{\overline{Y}}(\overline{y}; \theta)$ on kaksi kertaa jatkuvasti derivoituva $\forall y$
    \item $T = t(Y)$ tunnusluku, $\exists E_oT \forall \theta$
    \[
    \frac{\delta}{\delta \theta}E_0T = \int_{A}t(y)\frac{\delta}{\delta \theta} f_Y(y; \theta) dy
    \]
    \item 
    \[
    \frac{\delta^2}{\delta \theta^2}E_0 I = \int_A \frac{\delta^2}{\delta \theta^2}f_Y(y; \theta)dy
    \]
\end{enumerate}

Diskreetti malli:
\begin{enumerate}
    \item 
    \[
    \frac{\delta}{\delta \theta}E_{\theta}T = \sum_A t(y) \frac{\delta}{\delta \theta}f_Y(y; \theta)
    \]
    \item
    \[
    \frac{\delta^2}{\delta \theta^2}E_{\theta}I = \sum_A \frac{\delta^2}{\delta \theta^2} f_{\overline{Y}}(\overline{y}; \theta)
    \]
\end{enumerate}

Eksponenttiperheen jakaumat: Bernoulli, binomi, Poisson, normaali, gamma, eksponentti

Lause: Olkoon $f_{\overline{Y}}(\overline{y}; \theta)$ säännöllinen malli ja $l(\theta; y) = \log f_{Y}(y; \theta)$ sen log-uskottavuusfunktio. Tällöin
\begin{enumerate}
    \item $E[l'(\theta; Y)] = 0$
    \item $E[l'(\theta; Y)^2] = \textrm{var}[l'(\theta; Y)] = i(\theta)$
\end{enumerate}

\subsection{Harhattomuus}
\textbf{Unbiased, Harhaton, Väntevärdesriktig}

Määritelmä: Tilastollinen malli $f_Y(y; \theta)$, jonka parametri on $\Omega \in \mathcal{R}^d$. Funktion $g(\theta)$ estimaattori $T=t(Y)$ on harhaton, jos on voimassa 
\[
E_{\theta}(T) = g(\theta) \forall \theta \in \Omega
\]
Muussa tapauksessa estimaattori on harhaton ja sen harha $b(\theta) = E_{\theta}(T) - g(\theta)$. Estimoitavana itse parametri $\theta$: $E_{\theta}(T) = \theta$, $\forall \theta \in \Omega$, $b(\theta) = E_{\theta}(T) - \theta$

$d>1$: $E_{\theta}(T_j) = \theta_j$, kun $\theta \in \Omega$ ja $j = 1, \dots, d$

\begin{example}
$Y_1, \dots, Y_n$ jakauma, odotusarvo $\mu$. 

$\overline{Y} = \frac{1}{n}\sum_{i=1}^n Y_i$

$E(\overline{Y}) = \frac{1}{n}\sum_{i=1}^n E(Y_i) = \frac{1}{n}\sum_{i=1}^n \mu = \mu$
\end{example}

\begin{example}
$Y_1, \dots, Y_n \independent$ varianssi $\sigma^2$

$s^2 = \frac{1}{n-1}\sum_{i=1}^n (Y_i - Y)^2$, $E(s^2) = \sigma^2$ harhaton estimaattori

su-estimaattori $\hat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^n (Y_i - Y)^2$, $\hat{\sigma}^2 = (n - 1)s^2/n$, $E(\hat{\sigma}^2) = \frac{n-1}{n}\sigma^2$, $b(\hat{\sigma}^2) = -\frac{\sigma^2}{n}$
\end{example}

Harhattomuuden merkitys ja ongelmat: 
\begin{enumerate}
    \item Harhattomia estimaattoreita ei ole tai ne ovat epätyydyttäviä
    \item Ei ole invariantti parametrimuunniosten suhteen: Harhattomuus säilyy vain lineaarisissa muunnoksissa!
    \item Harhattomuus ei ole sopusoinnussa uskottavuusperiaatteen kanssa. Harhattomuus riippuu spesifioidusta mallista!
\end{enumerate}

\begin{example}
$Y_1, \dots, Y_n \sim \textrm{Exp}(1/\mu)$, $\hat{\mu} = \overline{Y}$ parametrin $\mu$ harhaton estimaattori

$\lambda = 1/\mu$ invarianssiominaisuus: $\hat{\mu} = \overline{Y}$, $S = Y_1 + \dots + Y_n \sim \sigma(n, \lambda)$

$f_S(s) = \frac{\lambda^n}{(n-1)!}s^{n-1}e^{-\lambda s}$, $s > 0$, $\hat{\lambda} = n/s$

$E(\hat{\lambda}) = \int_{0}^{\infty}\frac{n}{s}f_S(s)ds = n \int_0^{\infty} \frac{\lambda^n}{(n-1)!}s^{n-2}e^{-\lambda s}ds = \frac{n \lambda}{n - 1}\int_0^{\infty}\frac{\lambda^{n-1}}{(n -2)!}s^{n - 2}e^{-\lambda s}ds = \frac{n \lambda}{n -1}$

Harhainen! $T = (n-1)\hat{\lambda}/n = (n-1)/s$ harhaton!
\end{example}

Asymptoottinen harhattomuus: $T^{(n)}$ on funktion $g(\theta)$ estimaattori, joka perustuu kokoa $n$ olevaan aineistoon $(Y_1, \dots, Y_n)$. Jos estimaattorin harha lähestyy nollaa havaintojen kasvaessa
\[
\lim_{n \to \infty}E_{\theta}(T^{(n)}) = g(\theta), \forall \theta \in \Omega
\]
niin sanotaan, että $T^{(n)}$ tai oikeammion jono $T^{(1)}, T^{(2)}, \dots$ on asymptoottisesti harhaton.

\subsection{Momenttimenetelmä}

\vdots \\
m_d = \mu_d(\theta_1, \dots, \theta_d) \\

\subsection{Tehokkuus ja informaatioepäyhtälö}
Estimaattorin keskineliövirhe: $\overline{\theta}$:n komponentti tai itse parametri $\theta$. Jos $T$ on $g(\theta)$:n harhaton estimaattori, niin $T$:n saamat arvot ovat kesimäärin oikeita. Varianssi $\textrm{var}_{\theta}(T)$: kuinka tiiviisti todennäköisyysmassa on keskittynyt $g(\theta)$:n ympärille. 

Jos $T$ ei ole harhaton estimaattori, keskineliövirhe $E_{\theta}[(T - g(\theta))^2]$. $E_{\theta}[(T - g(\theta))^2] = \textrm{var}_{\theta}(T) + b(\theta)^2$, $b(\theta) = E_{\theta}(T) - g(\theta)$, $\textrm{var}_{\theta}(T) \geq \textrm{var}_{\theta}(T')$, $T$ tehokkaampi!

\begin{example}
Normaalimallin odotusarvo $Y_1, \dots, Y_n \sim N(\mu, \sigma_0^2) \independent$, $\sigma_0^2$ tunnettu luku

$\hat{\mu} = \overline{Y}$ on harhaton, $S=Y_1$ ja $T=(Y_1 + Y_2)/2$, varianssi $\sigma_0^2/n$, $\sigma_0^2$ ja $\sigma_0^2/2$, $\tilde{\mu}$ on tehokkain!
\end{example}

Informaatioepäyhtälö (d=1): Estimaattorin varianssilla on alaraja, jota ei voi ohittaa

Lause: Oletetaan, että $f_Y(y; \theta)$ on säännöllinen malli, ja olkoon $i(\theta)$ sen Fisherin informaatio. Olkoon lisäksi $T = t(Y)$ jokin funktion $g(\theta)$ estimaattori. 

Jos estimaattoirn $T$ harha on $b(\theta)$, niin 
\[
\textrm{var}_{\theta}(T) \geq \frac{[g'(\theta) + b'(\theta)]^2}{i(\theta)}
\]

Jos $T$ on $g(\theta)$:n harhaton estimaattori, niin 
\[
\textrm{var}_{\theta}(T) \geq \frac{g'(\theta)^2}{i(\theta)}
\]

ja erityisesti jos $T$ on $\theta$:n harhaton estimaattori $\textrm{var}_{\theta}(T) \geq \frac{1}{i(\theta)}$

Informaatioepäyhtälö: Cramér-Rao epäyhtälö

$T$:n tehokkuus $ = \frac{g'(\theta)^2/i(\theta)}{\textrm{var}_{\theta}(T)}100\%$, täystehokas = paras harhaton estimaattori

\begin{example}
Normaalimallin odotusarvo: $Y_1, \dots, Y_n \sim N(\mu, \sigma_0^2) \independent$, $\hat{\mu} = \overline{Y}$, $\textrm{var}\sigma_0^2/n$

$i(\mu) = n/\sigma_0^2$, $\frac{1}{i(\mu)} = \frac{1}{n/\sigma_0^2} = \frac{\sigma_0^2}{n}$, $\hat{\mu}$ on täystehokas
\end{example}

Informaatioepäyhtälö (d>1): $f_{\overline{Y}}(\overline{y}; \theta)$, parametri $\overline{\theta} = (\theta_1, \dots, \theta_d)$. Estimoitavana on jokin reaaliarvoinen funktio $g(\theta)$ parametrista $\overline{\theta}$. Fisherin informaatio on
\[
i_{a, b}(\theta) = E \bigg( - \frac{\delta^2}{\delta \theta_a \delta \theta_b} l(\theta; \overline{Y}) \bigg)
\]
\[
i^{-1}(\theta) = i(\theta)^{-1} = \begin{bmatrix}
i^{1,1}(\theta) & \dots & i^{1, d}(\theta) \\
\vdots & \ddots & \vdots \\
i^{d, 1}(\theta) & \dots & i^{d,d}(\theta) \\
\end{bmatrix}
\]
\[
\nabla g(\overline{\theta}) = \bigg( \frac{\delta}{\delta \theta_1}g(\overline{\theta}), \dots, \frac{\delta}{\delta \theta_d}g(\overline{\theta}) \bigg)
\]
Lause: Oletetaan, että $f_{\overline{Y}}(\overline{y}; \theta)$ on säännöllinen malli. Olkoon $T = t(Y)$ jokin funktion $g(\overline{\theta})$ estimaattori.
\begin{enumerate}
    \item Jos estimaattrorin harha on $b(\overline{\theta})$, niin
    \[
    \textrm{var}_{\theta}(T) \geq [\nabla g(\theta) + \nabla b(\theta)]' i^{-1}(\theta)[\nabla g(\theta) + \nabla b(\theta)]
    \]
    \item Jos $T$ on $g(\theta)$:n harhaton estimaattori, niin
    \[
    \textrm{var}_{\theta}(T) \geq [\nabla g(\theta)]' i^{-1}(\theta) [\nabla g(\theta)]
    \]
    \item Jos $T$ on $\theta_a$:n harhaton estimaattori $\textrm{var}_{\theta}(T) \geq i^{a,a}(\overline{\theta})$
\end{enumerate}

Voidaan osoittaa, että $i^{a, a}(\overline{\theta}) \geq \frac{1}{i_{a,a}(\overline{\theta})}$, jos $\theta_a$ on ortogonaalinen muihin parametreihin nähden. 

\begin{example}
$Y_1, \dots, Y_n \sim N(\mu, \sigma^2) \independent$
\[
i(\mu, \sigma^2) = \begin{bmatrix}
n/\sigma^2 & 0 \\
0 & n/(2\sigma^4) \\
\end{bmatrix}\textrm{   }i^{-1}(\mu, \sigma^2) = \begin{bmatrix}
\sigma^2/n & 0 \\
0 & 2\sigma^4/n \\
\end{bmatrix}
\]

$\textrm{var}(\hat{\mu} = \frac{\sigma^2}{n}$, täystehokas, $\textrm{var}(s^2) = \frac{2 \sigma^4}{n - 1}$, $s^2$:m tehokkuus on $\bigg(\frac{2\sigma^4}{n} / \frac{2 \sigma^4}{n - 1} \bigg) = \frac{n - 1}{n}$
\end{example}

Toinen muotoilu (d > 1): verrataan parametrivektorin estimaattoin kovarianssimatriisia Fisherin informaation käänteismatriisiin. Olkoon parametrina vektori $\overline{\theta} = (\theta_1, \dots, \theta_d)$, ja oletetaan $T = (T_1, \dots, T_d)$ on sen harhaton estimaattori. Satunnaisvektorin kovarianssimatriisilla $\textrm{cov}(T)$ tarkoitetaan symmetristä $d \times d$ -matriisia $(\sigma_{a, b})$, $\sigma_{a, b} = \textrm{cov}(T_a, T_b)$. Olkoon $a = (a_1, \dots, a_d)$ reaalisista vakioista koostuva vektori, $a'T = a_1T_1 + \dots + a_dT_d$. Odotusarvo $E(a'T) = a'\theta = a_1\theta_1 + \dots + a_d \theta_d$, varianssi on $\textrm{var}(a'T) = a'\textrm{cov}(T)a$, $a'\textrm{cov}(T)a \geq a'i^{-1}(\theta)a$, $a'[\textrm{cov}(T) - i^{-1}(\theta)]a \geq 0$. Eli näin ollen $\textrm{cov}(T) - i'(\theta) \geq 0$

\subsection{Tarkentuvuus}
\textbf{Consistency, Tarkentuvuus, Konsistens}

Asymptotiikka: estimaattorin käyttäytyminen, kun havaintojen lukumäärä kasvaa rajatta

Määritelmä: Malli $f_Y (y; \theta)$, estimoitavana $g(\theta)$. $T^{(n)} = t(Y_1, \dots, Y_n)$. Estimaattoria $T^{(n)}$ (jonoa $T^{(1)}, T^{(2)}, \dots$) sanotaan tarkentuvaksi, mikäli se suppenee stokastisesti kohti estimoitavan todellista arvoa eli lukua $g(\theta)$, ts. 
\[
\lim_{n \to \infty}P(|T^{(n)} - g(\overline{\theta})| > \epsilon ) = 0 \forall \epsilon > 0
\]

Heikko suurten lukujen laki: Olkoot $X_1, \dots, X_n$ riippumattomia satunnaismuuttujia, joilla on sama odotusarvo $\mu$ ja varianssi $\sigma^2$. Tällöin 
\[
\frac{1}{n}\sum_{i=1}^n X_i \to \mu, n \to \infty
\]

\begin{example}
$Y_1, \dots, Y_n \independent$, odotusarvo $\mu$, $\overline{Y} = \frac{1}{n}\sum_{i=1}^n Y_i \to \mu$, $\overline{Y}$ on $\mu$:n tarkentyva estimaattori
\end{example}

Lause: Oletetaan, että $g(\theta)$:n estimaattorille $T^{(n)}$ pätee $\forall \theta$.
\begin{enumerate}
    \item $\lim_{n \to \infty}E(T^{(n)}) = g(\theta)$ (asymptoottinen harhattomuus)
    \item $\lim_{n \to \infty}\textrm{var}(T^{(n)}) = 0$. 
\end{enumerate}
Tällöin $T^{(n)}$ on tarkentuva. 

\begin{example}
Normaalimallin varianssi $Y_1, \dots, Y_n \sim N(\mu, \sigma^2) \independent$. Varianssin $\sigma^2$ estimaattori $s^2 = \frac{1}{n-1}\sum_{i=1}^n(Y_i - \overline{Y})^2$

$s^2$ on harhaton, $\textrm{var}(s^2) = 2\sigma^4/(n-1) \to 0$, kun $n \to \infty$. 

$s^2$ on tarkentuva
\end{example}

Log-uskottavuusfunktio on $l(\overline{\theta}; \overline{y}) = \log f_{\overline{Y}}(\overline{y}; \overline{y}) = \sum_{i=1}^n \log f(y_i; \theta)$

Parametrin $\theta$ su-estimaattori: $\hat{\theta}^{(n)} = \hat{\theta}(Y_1, \dots, Y_n)$

Parametri $\theta$ on identifioituva: jos $\theta \neq \theta'$, niin $f(y; \theta)$ ja $f(y; \theta')$ ovat eri jakaumia.

Estimaattorin jakauma: $f_{\overline{Y}}(\overline{y}; \theta)$

\section{Tyhjentävyys}
\textbf{Sufficiency, Tyhjentävyys, Tillräcklighet}

\subsection{Tyhjentävä tunnusluku}
Tunnusluku: $y = (y_1, \dots, y_n)$ aineisto, mikä tahansa aineistosta $y$ laskettu suure tai funktio on tunnusluku $t=t(\overline{y}$ tai $t=(t_1(\overline{Y}), \dots, t_k(\overline{y}))$. 

Tyhjentävä tunnusluku:

Aineisto $\overline{y}$, satunnaisvektori $\overline{Y}$, tilastollinen malli yptf, ytf $f_{\overline{Y}}(\overline{y}; \overline{\theta})$.

Tunnusluku $t=t(y)$, $T = t(Y)$

Tunnusluku $T = t(Y)$ on parametrin $\theta$ tyhjentävä tunnusluku, jos satunnaisvektorin $Y$ ehdollinen jakauma ehdolla $T = t$ ei koskaan riipu $\theta$:sta eli $f_{Y|T}(y|t; \theta)$ ei sisällä $\theta$:aa

Jos aineistosta $y$ tiedetään sen verran, että tunnusluvun $t(y)$ arvo on $t$, niin aineiston tarkempi tuntemus ei enää tuo mitään lisäinformaatiota parametrista $\theta$.

Minimalinen tyhjentävä tunnusluku

\begin{example}
Toistokoemalli $\overline{y} = (y_1, \dots, y_n)$, $y_i = 0$ (ehjä) $y_i = 1$ (rikki)

$f_{\overline{Y}}(\overline{y}; \theta) = \theta^{\sum y_i}(1 - \theta)^{n - \sum y_i}$, $K = \sum_{i=1}^n Y_i$, $K \sim \textrm{Bin}(n, \theta)$

$f_K(k; \theta) = {n \choose k}\theta^k (1 - \theta)^{n-k}$

$f_{Y|K}(y|k; \theta) = \frac{P(Y = y, K = k)}{P(K = k)}$

$\sum y_i = k, \{Y = y\}$ sisältyy tapahtumaan $\{K = k \}$

\[
f_{Y|K}(y|k; \theta) = \frac{P(Y = y)}{P(K = k)} = \frac{\theta^{\sum y_i}(1 - \theta)^{n - \sum y_i}}{{n \choose k}\theta^k (1 - \theta)^{n - k}} = \frac{1}{{n \choose k}}
\]
\end{example}

\subsection{Faktorointikriteeri}
Lause: Tunnusluku $T = t(Y)$ on parametrin $\theta$ tyhjentävä tunnusluku jos ja vain jos yptf, ytf $f_{\overline{Y}}(\overline{y}; \overline{y})$ voidaan kirjoittaa $f_{\overline{Y}}(\overline{y}; \overline{\theta}) = h(\overline{y})g(t(\overline{y}; \overline{\theta})$ $\forall y, \theta \in \Omega$

Tunnusluku $T = t(\overline{Y})$ on tyhjentävä jos ja vain jos mallin uskottavuusfunktio voidaan valita siten, että se riippuu aineistosta vain $t(\overline{y})$:n välityksellä.

\begin{example}
Normaalimalli. Olkoot $Y_1, \dots, Y_n \sim N(\mu, \sigma^2) \independent$
\[
L(\mu, \sigma^2) = (\sigma^2)^{-n/2}\exp \bigg( - \frac{(n-1)s^2 + n(\overline{y} - \mu)^2}{2\sigma^2} \bigg) 
\]
$(\overline{y}, s^2)$ on tyhjentävä tunnusluku, $\sigma^2 = \sigma_0^2 > 0$
\[
l(\mu) = \exp \bigg( - \frac{n(\overline{y} - \mu)^2}{2 \sigma_0^2}
\]
$\overline{y}$ on $\mu$:n tyhjentävä tunnusluku
\end{example}

Eksponenttiperheen mallit: havaintojen lukumäärästä $n$ riippumatta sillä on tyhjentävä tunnusluku, dimensio sama kuin mallin parametrin

\section{Hypoteesin testaus}
\subsection{Johdanto}
Testin vaiheet yleisesti: parametrinen tilastollinen malli, testin vaiheet
\begin{enumerate}
    \item Asetetaan nollahypoteesti $H_0$ ja vastahypoteesi $H_1$
    \item Valitaan käytettävä testisuure
    \item Lasketaan testisuureen arvot, merkitsevyystaso ja p-arvo
    \item Tehdään johtopäätökset
\end{enumerate}

\subsection{Peruskäsitteet ja testin suorittaminen}
Hypoteesit ja kysymystenasettelu: Hypoteesi on väite, joka koskee mallin parametria ja voidaan kirjoitaa muotoon $\theta \in \Omega'$, $\Omega' \in \Omega$. 

Hypoteesi on yksinkertainen: yksi piste, muuten yhdistetty

Nollahypoteesi $H_0$: $\theta \in \Omega_0$

Kategorinen päätös: nollahypoteesi joko hyväksytään tai hylätään

$H_1: \theta \in Omega_1$, $\Omega_0 \cup \Omega_1 = \Omega$

Testisuure: mitataan nollahypoteesin ja ainesiton välistä yhteensopivuutta $t = t(\overline{y})$, monotonisuusominaisuus: pienet $t$:n arvot viittaavat siihen, että aineisto on sopusoinnussa $H_0$:n kanssa

Testisuureen valinta: malli, $H_0$ ja $H_1$

Havaittu merkitsevyystaso:

Oletetaan, että testaaminen on päätetty perustaa testiruureeseen $t$, jonka suuret arvot ovat nollahypoteesille $H_0$ kriittisiä. Olkoon $T = t(Y)$ sitä vastaava satunnaismuuttuja, $\overline{y}$ havaittu aineisto.
\[
p = p(y) = P_{\theta}(T \geq t(y)), \theta \in \Omega_0
\]
jos $H_0$ on yksinkertainen tai jos tämä todennäköisyys on sama $\forall \theta \in \Omega_0$.
\[
p = p(y) = \textrm{sup}_{\theta \in \Omega}P_{\theta}(T \geq t(y))
\]
Nollahypoteesin pätiessä laskettu yläraja sille todennäköisyydelle, että satunnaismuuttuja $T$ saa arvon, joka on yhtä suuri tai suurempi kuin $t(y)$.

\begin{example}
Lantin harhattomuus $H_0: \theta = 1/2$

$K \sim \textrm{Bin}(1000, \theta) \to K \sim \textrm{Bin}(1000, 1/2)$

$E_{H_0}(K) = 500$, $p=P_{H_0}\{|K-500| \geq |k - 500| \} = P_{H_0}(K \leq 500 - |k - 500| + P_{H_0}(K \geq 500 + |k - 500|\}$
\end{example}

\subsection{Havaitun merkitsevyystason tulkinnasta}
P-arvo on tunnusluku: p-arvo on aineistosta laskettu tunnusluku $p = p(\overline{y})$.

Päätösteoreettinen lähestymistapa: Joko $H_0$ hyväksytään, tai hylätään ja $H_1$ hyväksytään. $\alpha \in (0,1)$ merkitsevyystao $p > \alpha \to H_0$ hyväksytään, $p \leq \alpha$, $H_0$ hylätään. 

Melkein merkitsevä, merkitsevä, erittäin merkitsevä

$\alpha$ on yläraja hylkäämisvirheen riskille eli todennäköisyydelle. Valitaan testi niin, että hyävksymisvirheen riski on mahdollisimman pieni!

p-arvo ei ole todennäköisyys sille, että $H_0$ pätee. Frekvenistisessä päättelyssä satunnaisuutta ei liitetä parametriin tai hypoteesiin. Pieni p-arvo: on saatu sellainen aineisto, joka on nollahypoteesin pätiessä epätodennäköinen. 

Valintakorjaus:

Ennen aineistoon tutustumista päätetään, millaisia hypoteeseja halutaan tutkia ja mitä testejä tähän tarkoitukseen käytetään. 

$H_0: \theta \in \Omega_0$ suoritetaan $k$ testiä, p-arvot $p_j(\overline{y})$, $j = 1, \dots, k$. $Q = \textrm{min}(P_1, \dots, P_k)$
\[P_{\theta}(Q \leq q) = 1 - P_{\theta}(Q > q)\]
\[1 - P_{\theta}(P_1 > q, \dots, P_k < q)\]
\[1 - P_{\theta}(P_1 > q)\dots P_{\theta}(P_k > q)\]
\[\leq 1 - (1 - q)^k\]

Testit ovat riippumattomia, vielä karkeampi arvio
\[
P_{\theta}(Q \leq q) = P_{\theta}\bigg( \bigcup_{j=1}^k (P_j \leq q) \bigg) \leq \sum_{j=1}^k P_{\theta}(P_j \leq q) \leq kq
\]
Bonferroni-korjattu p-arvo!

\subsection{Normaalimallin perustestit}
$\sigma^2 = \sigma_0^2 > 0$, $H_0: \mu = \mu_0$, $H_1: \mu \neq \mu_0$, $z = \sqrt{n})(\overline{y} - \mu_0)/\sigma_0$

Kaksisuuntainen testi: $p=P_{\mu_0}(|Z| \geq |z|) = 2[1 - \phi(|z|)]$

Yksisuuntainen testi: $H_0: \mu = \mu_0$, $H_1: \mu > \mu_0$, $p = P_{\mu_0}(Z \geq z) = 1 - \phi(z)$

Varianssi on kiusaparametri: $Z = \sqrt{n}(\overline{Y} - \mu_0)/\sigma$

$S^2 = \frac{1}{n-1}\sum_{i=1}^n (Y_i - \overline{Y})^2$, $T = \frac{\overline{Y} - \mu_0}{S/\sqrt{n}}$ noudattaa Studentin t-jakaumaa, vapausaste $n-1$

Varianssin testi: $H_0: \sigma^2 = \sigma_0^2$, $H_1: \sigma^2 > \sigma_0^2$
\[
(n-1)s^2/\sigma^2 \sim \chi_{n-1}^2
\]
p-arvoksi saadaan vastaava oikeanpuoleinen häntätodennäköisyys $\chi_{n-1}^2$-jakaukmasta

\subsection{Testin voima ja Neyman-Pearson-teoria}
Kriittiset alueet ja voimafunktio: $f_{\overline{Y}}(\overline{y}; \theta)$, $\Omega$, $H_0: \theta \in \Omega_0$, $H_1: \theta \in \Omega_1$

Testisuure $t$, p-arvo $p$ $C_{\alpha}$ on niiden aineistojen $y$ joukko, joiden havaitseminen johtaa $H_0$:n hylkäämiseen, ts. $C_{\alpha} = \{y; p(y) \leq \alpha \}$

Testisuureen $t$ indusoima $\alpha$-tasoinen kriittinen alue
\[
P_{\theta}(Y \in C_{\alpha}) = P_{\theta}({H_0}) \leq \alpha, \theta \in \Omega_0
\]
Testisuureen $\alpha$-tasoinen voima tai voimafunktio on
\[

\end{cases}
\]

\end{example}

\subsubsection{Testien voiman vertailu ja Neyman-Pearson-apulause}

$t$ ja $t'$ ovat kaksi testisuuretta, joita voidaan käyttää samojen hypoteesien testaukseen. $\pi_{\alpha}(\theta; t)$ ja $\pi_{\alpha}(\theta: t')$
\begin{enumerate}
    \item $t$ on voimakkaampi kuin $t'$ pisteessä $\theta \in \Omega_1$, jos $\pi_{\alpha}(\theta; t) \geq \pi_{\alpha}(\theta; t')$
    \item $t$ on tasaisesti voimakkaampi kuin $t'$, jos $\forall \theta \in \Omega_1$ pätee $\pi_{\alpha}(\theta; t) \geq \pi_{\alpha}(\theta; t')$
    \item $t$ on voimakkain testisuure, jos $\forall \theta \in \Omega_1$, $\forall t'$ pätee 1)
    \item $t$ on tasaisesti voimakkain testisuure, jos $\forall \theta \in \Omega_1$, $\forall t'$ pätee a)
\end{enumerate}

Neyman-Pearson apulause: tarkastellaan tilastollista mallia $f_Y(y; \theta)$ ja yksinkertaisia hypoteeseja $H_0: \theta: \theta_0$ ja $H_1: \theta; \theta_1$. 
\[
v(\overline{y}) = \frac{L(\overline{\theta}_1; \overline{y})}{L(\overline{\theta}_0; \overline{y})} = \frac{f_Y(\overline{y}; \overline{\theta)_1}}{f_Y(\overline{y}; \overline{\theta}_0}
\]
Tällöin $v$ on voimakkain testisuure hypoteesille jokaisella sellaisella merkitsevyystasolla $\alpha(0, 1)$, jolle pätee $P_{\theta_0}(v(\overline{Y}) \geq v_{\alpha}) = \alpha$ jollakin $v_{\alpha}$.

\begin{example}
Normaalimallin odotusarvo, uskottavuusosamäärän testi

$Y_1, \dots, Y_n \sim N(\mu, \sigma_0^2) \independent$

$L(\mu; \overline{y}) = \exp \bigg( - \frac{n(\overline{y} - \mu)^2}{2 \sigma_0^2} \bigg)$
\[
\frac{L(\mu_1; \overline{y}}{L(\mu_0; \overline{y}} = \exp \bigg( \frac{n}{2  \sigma_0^2} [2 \overline{y}(\mu - \mu_0) - (\mu_1^2 - \mu_0^2)] \bigg), \overline{y} \geq c', z \geq c''
\]
$z$-testisuure on voimakkain testisuure!
\end{example}

Monotonine uskottavuusosamäärä $f_{\overline{Y}}(\overline{y}; \theta)$
\[
v(\overline{y}) = \frac{L(\theta'; \overline{y})}{L(\theta; \overline{y}} = \frac{f_{overline{Y}}(\overline{y}; \theta')}{f_{\overline{Y}}(\overline{y}; \theta)}
\]
riippuu aineistosta vain $t(\overline{y})$:n välityksellä, aidosti kasvava $\forall \theta, \theta' \in \Omega$, $\theta < \theta'$. $t(\overline{y})$ on tasaisesti voimakkain testisuure.

\begin{example}
Eksponenttimalli, $Y_1, \dots, Y_n \sim \textrm{Exp}(\lambda)\independent$, $f_{\overline{Y}}(\overline{y}; \lambda) = \lambda^n \exp( - \lambda \sum_{i=1}^n y_i)$, $0 < \lambda < \lambda'$
\[
\frac{f_{\overline{Y}}(\overline{y}; \lambda')}{f_{\overline{Y}}(\overline{y}; \lambda)} = \bigg( \frac{\lambda'}{\lambda} \bigg)^n \exp ( - (\lambda' - \lambda) \sum_{i=1}^n y_i), \lambda' - \lambda > 0
\]
tunnusluvun $-\sum_{i=1}^n y_i$ aidosti kasvava funktio. Mallilla on siis monotoninen uskottavuusosamäärä. Tasaisesti voimakkain testisuure $t(\overline{y}) = \sum_{i=1}^n y_i$, $H_0: \lambda = \lambda_0$, $H_1: \lambda > \lambda_0$.

$p=P_{\lambda_0}(T \leq t(\overline{y}))$, $T = \sum_{i=1}^n Y-i$
\end{example}

\subsection{Uskottavuusfunktioon perustuvia testejä I}

\end{enumerate}

Uskottaviuusosamäärän testisuure:

r(\overline{y}) = 2 \log \frac{L(\hat{\theta}; \overline{y})}{L(\theta_0; y)} = 2[l(\hat{\theta}; y) - l(\theta_0; \overline{y})]

\end{example}

Waldin testisuure: $\hat{\theta} - \theta_0$, $w^{1/2}(y) = \sqrt{i(\theta_0)}(\hat{\theta} - \theta_0)$, $w(\overline{y}) = i(\theta_0)(\hat{\theta} - \theta_0)^2$

\subsection{Uskottavuusfunktioon persutuvia testejä II}
Johdanto: mallit, parametrit, useampiulotteinen tai nollahypoteesi yhdistetty. $f_{\overline{Y}}(\overline{y}; \theta)$, $\Omega \subset \mathcal{R}^d$, $H_0: \theta \in \Omega_0$. Tarkasteltavana kaksi mallia: vapaa malli $f_{\overline{Y}}(\overline{y}; \theta)$, $\theta \in \Omega$, rajoitettu malli $f_{\overline{Y}}(\overline{y}; \overline{\theta})$, $\theta \in \Omega_0$. Vapaa su-estimaatti $\hat{\theta} \in \Omega$ ja rajoitettu su-estimaatti $\hat{\theta}_0 \in \Omega_0$. $L(\hat{\theta}; \overline{y}) = \textrm{max}L(\overline{\theta}; \overline{y})$, $L(\hat{\theta}_0; \overline{y}) = \textrm{max} L(\overline{\theta}; \overline{y})$

Uskottavuusosamäärän testisuure:
\[
r(\overline{y}) = 2 \log \frac{L(\hat{\theta}; \overline{y})}{L(\hat{\theta}_0; \overline{y})} = 2 [ l(\hat{\theta}; \overline{y}) - l(\hat{\theta}_0; \overline{y})]
\]

Oletukset:
\begin{enumerate}
    \item Mallin parametri voidaan osittaa muotoon $\theta = (\psi, \lambda)$, jossa $\psi = (\theta_1, \dots, \theta_q)$ ja $\lambda = (\theta_{q+1}, \dots, \theta_d)$
    \item Parametriavaruus $\Omega$ voidaan kirjoittaa $\Omega = \Omega' \times \Omega''$, $\Omega' \subset \mathcal{R}^q$, $\Omega'' \subset \mathcal{R}^{d-q}$
    \item Nollahypoteesi: $H_0: \psi = \psi_0$, $\psi_0 \in \Omega'$, $\Omega_0 = \{ \psi_0 \} \times \Omega'' = \{(\psi_0, \lambda): \lambda \in \Omega'' \}$, su-estimaatti $\hat{\theta}_0 = (\psi_0, \hat{\lambda_0})$, $L(\psi_0, \hat{\lambda}_0; \overline{y}) = \textrm{max}_{\lambda \in \Omega''}L(\psi_0, \lambda: \overline{y})$
\end{enumerate}

Behrensin ja Fisherin ongelma: difference between means of two normally distributed populations when the variances of the two populations are not assumed to be equal

Uskottavuudosamäärän testisuure: $r(\overline{y}) = 2[l(\hat{\theta}; \overline{y}) - l(\hat{\theta}_0; \overline{y})]$, $r(\overline{Y}) \sim \chi_q^2$, $p \approx P(\chi_q^2 \geq r(\overline{y})$, q = kiinnostavien parametrien dimensio

Waldin testisuure: $\hat{\theta} \sim N-d(\theta, i^{-1}(\theta))$
\[
i(\theta) = \begin{bmatrix}

\vdots & \ddots & \vdots \\
i_{d, 1}(\theta) & \dots & i_{d,d}(\theta) \\
\end{bmatrix}, i_{a,b}(\theta) = E\bigg( - \frac{\delta^2}{\delta \theta_a \delta \theta_b}l(\overline{\theta}; \overline{Y}) \bigg), i^{-1}(\theta) = \begin{bmatrix}
i^{\psi, \psi}(\theta) & i^{\psi, \lambda}(\theta) \\
i^{\lambda, \psi}(\theta) & i^{\lambda, \lambda}(\theta) \\
\end{bmatrix}

$\hat{\theta} = (\hat{\psi}, \hat{\lambda})$, $\hat{\psi} \sim N_q(\psi, i^{\psi, \psi}(\theta)$, $(\hat{\psi} - \psi)'i^{\psi, \psi}(\theta)^{-1}(\hat{\psi} - \psi) \sim \chi_q^2$

$w/y) = (\hat{\psi} - \psi_0)'i^{\psi, \psi}(\hat{\theta})^{-1}(\hat{\psi} - \psi_0)$, $\hat{\theta} = (\hat{\psi}, \hat{\lambda})$, $w(\overline{Y}) \sim \chi_q^2$

Raon testisuure: $\nabla l(\overline{\theta}, \overline{y}) = \bigg(\frac{\delta}{\delta \theta_1} l(\overline{y}; \overline{y}), \dots, \frac{\delta}{\delta \theta_d} l(\overline{\theta}; \overline{y}) \bigg)$

$\nabla l(\overline{\theta}; \overline{y}) = (\nabla_{\psi}l(\overline{\theta}; \overline{y}), \nabla_{\lambda}l(\overline{\theta}; \overline{y})$

$u(\overline{y}) = [\nabla_{\psi}l(\hat{\theta}_0; \overline{y})]'i^{\psi, \psi}(\hat{\theta}_0)[\nabla_{\psi}l(\hat{\theta}_0; \overline{y})]$

$u(\overline{Y}) \sim \chi_q^2$

\[
i(\mu, \sigma^2) = \begin{bmatrix}

q=1: Luottamusväli parametrin yhdelle komponentille
\[
\{\theta_1: (\hat{\theta}_1 - \theta_1)^2/i^{1,1}(\hat{\theta}) < \chi_1^2(\alpha) \}
\]
\[
(\hat{\theta_1} - z_{\alpha/2}i^{1,1}(\hat{\theta}^{1/2}, \hat{\theta}_1 + z_{\alpha/2}i^{1,1}(\hat{\theta})^{1/2}))
\]

q=2: Luottamusellipsi parametrin komponenttien parille
$(\theta_1, \theta_2)$ luottamusjoukko $\to$ ellipsi, keskipiste $(\hat{\theta}_1, \hat{\theta}_2)$
\[
i^{\psi, \psi}(\hat{\theta}) = \begin{bmatrix}
i^{1,1}(\hat{\theta}) & i^{1,2}(\hat{\theta}) \\
i^{2,1}(\hat{\theta}) & i^{2,2}(\hat{\theta}) \\
\end{bmatrix}
\]
\[
\textrm{cor}(\hat{\theta_1}, \hat{\theta}_2) = \frac{i^{1,2}(\hat{\theta)}}{i^{1,1}(\hat{\theta})^{1/2}i^{2,2}(\hat{\theta})^{1/2}}
