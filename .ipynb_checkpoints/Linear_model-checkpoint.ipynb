\subsection{Matriisitulo}
Matriisien $\mathbf{A}$ $(n \times m)$ ja $\mathbf{B}$ $(m \times l)$ tulo on $\mathbf{AB} = \bigg[ \sum_{k=1}^m a_{ik}b_{kj} \bigg] (n \times l)$

$(\mathbf{AB})\mathbf{C} = \mathbf{A}(\mathbf{BC})$

$(\mathbf{AB})\mathbf{C} = \mathbf{ABC}$ 

$\mathbf{A}(\mathbf{B} + \mathbf{C}) = \mathbf{AB} + \mathbf{AC}$

$(\mathbf{A} + \mathbf{B})\mathbf{C} = \mathbf{AC} + \mathbf{BC}$

\subsection{Transponointi}
\textbf{Transpose, Transpoosi, Transponat}
Matriisin $\mathbf{A}$ $(n \times m)$ transpoosi $\mathbf{A}'$ tai $\mathbf{A}^T$ $m \times n$ on $\mathbf{A}' = [a_{ij}]' = [a_{ji}]$

\textbf{Symmetric, Symmetrinen, Symmetrisk}
Symmetrinen matriisi: $\mathbf{A} = \mathbf{A}'$

$(\mathbf{A}')' = \mathbf{A}$

$(\mathbf{A} + \mathbf{B})' = \mathbf{A}' + \mathbf{B}'$

$(\mathbf{AB})' = \mathbf{B}' \mathbf{A}'$

$(\mathbf{ABC})' = \mathbf{C}' \mathbf{B}' \mathbf{A}'$ 

$(\lambda \mathbf{A})' = \lambda \mathbf{A}'$

$(\mathbf{x}, \mathbf{y}) = \mathbf{x}'\mathbf{y}$ 

$||\mathbf{x}|| = \sqrt{(\mathbf{x}'\mathbf{x})}$ 

\subsection{Käänteismatriisi}
$\mathbf{A}, (n \times n)$, $\mathbf{AB} = \mathbf{I}_n$ ja $\mathbf{BA} = \mathbf{I}_n$, $\mathbf{B} = \mathbf{A}^{-1}$
A on täysiasteinen = epäsingulaarinen = kääntyvä

\textbf{Nonsingular, epäsingulaarinen, icke-singulär}

\textbf{Invertible, Kääntyvä, inverterbar}

Ortogonaalinen matriisi on epäsingulaarinen: $\mathbf{A}^{-1} = \mathbf{A}'$ , $\mathbf{A}'\mathbf{A} = \mathbf{I}_n$ ja $\mathbf{A}\mathbf{A}' = \mathbf{I}_n$

Epäsingulaariset matriisit: $(\mathbf{A}^{-1})^{-1} = \mathbf{A}$, $(\mathbf{AB})^{-1} = \mathbf{B}^{-1}\mathbf{A}^{-1}$ ja $(\mathbf{A}')^{-1} = (\mathbf{A}^{-1})'$ 

\subsection{Ominaisarvot ja vektorit}
$\mathbf{A} (n \times n)$ ominaisarvot $\lambda_1, \dots, \lambda_n$ ovat $\textrm{det}(\mathbf{A} - \lambda I_n) = 0$ ratkaisut

$\mathbf{Au}_i = \lambda \mathbf{u}_i$ ja $(\mathbf{A} - \lambda_i \mathbf{I}_n)u_i = 0$ 

Pääakseliesitys: $\mathbf{A} = \mathbf{U\Lambda U}'$  jossa $\Lambda = \textrm{diag}[\lambda_1 \dots \lambda_n]$ ja $\mathbf{U} = [\mathbf{u}_1 \dots \mathbf{u}_n]$ 

\textbf{Eigendecomposition, pääakseliesitys, diagonalisering}

Singulaariarvohajotelma: $\mathbf{A} = \mathbf{RDO}'$ jossa $\mathbf{D} = \textrm{diag}[d_1 \dots d_m] \geq 0$ 

\textbf{Singulaariarvohajotelma, Singulärvärdesuppdelning}

$\mathbf{R}'\mathbf{R} = \mathbf{I}_m$ ortonormaalit sarakkeet $\mathbf{O}'\mathbf{O} = \mathbf{I}_m = \mathbf{OO}'$

$d_1 \geq \dots d_m \geq 0$ ovat matriisin $\mathbf{A}'\mathbf{A}$ ominaisarvojen neliöjuuret

$\mathbf{O}$:n sarakkeet ovat matriisin $\mathbf{A}'\mathbf{A}$ ominaisvektorit

\subsection{Matriisin jälki}
\textbf{Trace, jälki, spår}
Diagonaalialkioiden summa eli $\textrm{tr}(\mathbf{A}) = \sum_{i=1}^n a_{ii}$ 

$\mathrm{tr}(\mathbf{A} + \mathbf{B}) = \textrm{tr}(\mathbf{A}) + \textrm{tr}(\mathbf{B})$ ja $\textrm{tr}(\mathbf{AB}) = \textrm{tr}(\mathbf{BA})$ ja $\textrm{tr}(\mathbf{A}) = \sum_{i=1}^n\lambda_i$ , jossa $\lambda_1 \dots \lambda_n$ ovat $\mathbf{A}$:n ominaisarvot

\subsection{Matriisin determinantti}
\textbf{Determinant, determinantti, determinant}
$\mathbf{A} (n \times n)$ determinantti $\textrm{det}(\mathbf{A}) = |\mathbf{A}|$ 

$\textrm{det}(\mathbf{AB}) = \textrm{det}(\mathbf{A})\textrm{det}(\mathbf{B})$ ja $\textrm{det}(\mathbf{A}^{-1}) = \frac{1}{\textrm{det}(\mathbf{A}}$  ja $\textrm{det}(\mathbf{A'}) = \textrm{det}(\mathbf{A})$ 

Jos $\textrm{det}(\mathbf{A}) = 1$ tai $-1$, niin $\mathbf{A}$ on ortogonaalinen
