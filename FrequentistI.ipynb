\section{Havaintojen mallintaminen}
\subsection{Havaintoja vastaava todennäköisyysmalli}
Aineisto: $y_1, \dots, y_n$, $n=$ otoskoko

Arvot $y_1, \dots, y_n$ ovat satunnaismuuttujien $Y_1, \dots, Y_n$ toteutuneita arvoja

$y_1 = Y_1(\omega^{act)}$, $y_2 = Y_2(\omega^{act})$, $\dots, y_n = Y_n(\omega^{act})$

$\overline{y}=(y_1, \dots, y_n)$, $\overline{Y}=(Y_1, \dots, Y_n)$ vektorit, $\overline{y}=\overline{Y}(\omega^{act})$

Tilastollisen päättelyn tavoitteena on tehdä aineiston perusteella johtopäätöksiä $Y$:n jakaumasta

$Y$:n jakaumaa mallinnetaan parametrisella mallilla, $ \hat{\theta} = (\theta_1, \dots, \theta_n)$

Parametrin arvo on kiinnitetty: $\hat{Y} = f(\hat{y}; \theta) = f(y_1, \dots, y_n ; \theta)$ = yhteispistetodennäköisyysfunktio tai yhteistiheysfunktio

Satunnaismuuttujat ovat riippumattomia: $f(\hat{y}; \theta) = f_{Y_1}(y_1; \theta)f_{Y_2}(y_2; \theta) \dots = \prod_{i=1}^n f_{Y_i}(y_i ; \theta)$

$Y_1, \dots, Y_n$ ovat satunnaisotots jakaumasta, $\Theta \in \mathcal{R}$ parametriavaruus

$P_{\theta}(Y_i = 1) = \theta/N$, $P_{\theta}(Y_i = 0) = 1 - \theta/N$

P_{\theta}(Y_i = y_i) = \bigg( \frac{\theta}{N} \bigg)^{y_i} \bigg(1 - \frac{\theta}{N} \bigg)^{1 - y_i}, y_i = 0, 1 \to f_{Y_i}(y_i; \theta)
\]

\[
f(\overline{y}; \theta) = f(y_1, \dots, y_n; \theta) = f_{Y_1}(y_1; \theta)f_{Y_2}(y_2; \theta) \dots f_{Y_n}(y_n; \theta) = 
\]

\]
\[

\subsection{Nasta purkissa}
$Y_i = \begin{cases}
1 \\
0
\end{cases}$

$f(\overline{y}; \theta) = \theta^{y_1}(1 - \theta)^{1 - y_i} \dots \theta^{y_n} (1 - \theta)^{1 - y_n} = \theta^{t(y)} (1 - \theta)^{n - t(y)}$, $t(y) = \sum_{i=1}^n y_i$

\subsection{Binomikoe}
Bernoullin koe= 2 tulosmahdollisuutta

$Y_1, \dots, Y_n$ yhteisjakaumalla on yhteispistetodennäköisyysfunktio $f(\overline{y}; p) = p^{y_1}(1-p)^{1 - y_1}\dots = p^{t(y)}(1 - p)^{n - t(y)}$, $t(y) = \sum_{i=1}^n y_i$

$X = t(Y) = \sum_{i=1}^n Y_i$

$X \sim \textrm{Bin}(n, p)$

$P_p(X = x) = {n \choose x} p^x (1-p)^{n - x}, x=0,1, \dots, n$

\subsection{Kaksi lähestymistapaa päättelyyn}
Frekvenistinen lähestymistapa: $\theta$ on tuntematon, mutta kiinteä $\{f(\overline{y}; \theta): \theta \in \Theta \}$ jakaumaperhe

\section{Estimointiteoriaa}
$\{ f(\overline{y}; \theta), \theta \in \Theta)$, $\hat{\theta}$ paras arvaus parametrin todelliselle arvolle, $\overline{y} \to f(\overline{y}; \hat{\theta})$

\subsection{Parametri ja tunnusluku}
Parametri: mikä tahansa vektorin $Y$ jakauman ominaisuutta kuvaava luku

$EY_i = p$, $\textrm{var}Y_i = p(1-p)$, $p=\theta/N$

Parametri $t=k(\theta)$

Tunnusluku, statistic: voidaan laskea ilman parametria, $t(\overline{y})$

\subsection{Estimaatti, estimaattori ja otantajakauma}
Nasta purkissa, $\hat{\theta} = \frac{1}{n}\sum_{i=1}^n y_i$, $\hat{\theta} \in \Theta$, $\hat{t} \in \{ k(\theta): \theta \in \Theta \}$, minimaainen vaatimus estimaatille!

Satunnaismuuttujan $t(Y)$ jakauma: tunnusluvun otantajakauma, sampelfördelning

Suuren $h(Y, \theta)$ jakauma: suuren otantajakauma

Estimaattori: satunnaismuuttuja $t(\overline{Y})$

Nasta purkissa: estimaatti $\hat{\theta} = \frac{1}{n} \sum_{i=1}^n y_i$

Estimaattori: $\hat{\theta}(\overline{Y}) = \frac{1}{n}\sum_{i=1}^n Y_i$

"Parametrin estimaatin todennäköisyysjakauma"

\subsection{Todennäköisyyslaskenta}
$\textrm{var}(X) = E(X - \mu)^2$, $E(aX + b) = aEX + b$, $\textrm{var}(aX + b) = a^2 \textrm{var}X$

$E(X_1 + X_2) = EX_1 + EX_2$, $\textrm{var}(X_1 + X_2) = \textrm{var}X_1 + \textrm{var}X_2$

$E(X - a)^2 = E(X - \mu)^2 + (\mu - a)^2 = \textrm{var}X + (\mu - a)^2$

\subsection{Otantajakauman ominaisuuksia}
$E_{\theta}[\hat{\theta}(Y)] = \frac{1}{n}\sum_{i=1}^n E_{\theta}(Y_i) = \frac{1}{n}n\theta = \theta$

$\textrm{var}_{\theta}[\hat{\theta}(Y)] = \frac{1}{n^2}\sum_{i=1}^n \textrm{var}_{\theta}(Y_i) = \frac{1}{n^2}n\theta(1 - \theta)$

Jos $E_{\theta}[\hat{\theta}(Y)] = \theta$, $\forall \theta$, estimaattori $\hat{\theta}(Y)$ on harhaton (unbiased)

\textbf{Mean unbiased, Harhaton odotusarvon mielessä, Väntevärdesriktig}

\textbf{Median unbiased, Harhaton mediaanin mielessä, Medianriktig}

$\textrm{bias}(\hat{\theta}) = E\hat{\theta} - \theta$

Keskineliövirhe, mean squared error: $mse_{\theta}(\hat{\theta}|Y) = E_{\theta}[(\hat{\theta}(Y) - \theta)^2]$, $mse(\hat{\theta}) = E(\hat{\theta} - \theta)^2$

$mse_{\theta}(\hat{\theta}|Y) = \textrm{var}_{\theta}(\hat{\theta} | Y) + \bigg( \textrm{bias}_{\theta}(\hat{\theta}(Y) \bigg)^2$

Keskineliövirheen neliöjuuri, root mean squared error: $rmse_{\theta}(\hat{\theta}(\overline{Y})) = \sqrt{mse_{\theta}\hat{\theta}(Y)}$

Keskivirhe, standard error, standardfel = keskineliövirheen neliöjuuri

\subsection{Tarkentuvuus}
Parametrin $\theta$ estimaattori $\hat{\theta}(Y)$ on tarkentuva, mikäli se suppenee kohti parametrin $\theta$ todellista arvoa otoskoon $n$ kasvaessa rajatta

Stokastinen suppeneminen: Jono $X_1, X_2, \dots$ suppenee stokastisesti kohti $a$
\[
P(|X_n - a| > \epsilon) \to 0 \forall \epsilon > 0
\]

Tarkentuvuus: Jos $\hat{\theta}(\overline{Y}) \to \theta$ $\forall \theta \in \Theta$, kun otoskoko $n$ ja sen mukana havaintosatunnaisvektorin $\overline{Y}=(Y_1, \dots, Y_n)$ pituus kasvavat rajatta, niin tällöin sanotaan, että estimaattori $\hat{\theta}(\overline{Y})$ on tarkentuva.
\[
\textrm{mse}_{\theta}(\hat{\theta}(Y)) = \textrm{var}_{\theta}(\hat{\theta}(\overline{Y})) = \frac{1}{n}\theta (1 - \theta), n \to \theta, \tp 0 \forall 0 \leq \theta \leq 1
\]

\section{Suurimman uskottavuuden menetelmä ja momenttimenetelmä}
\subsection{Uskottavuusfunktio}
\textbf{Likelihood funciton, Uskottavuusfunktio, Likelihood-funktion}

$\overline{y} \to f(\overline{y}; \theta)$ on tiheysfunktio

$\theta \to f(\overline{y}; \theta)$ on uskottavuusfunktio

$L(\theta) = f(\overline{y}; \theta)$

\begin{example}
Pallot kulhossa, $y = (1, 0, 0, 0, 1, 0, 0)$,
\[
L(\theta) = \bigg( \frac{\theta}{5} \bigg)^2 \bigg(1 - \frac{\theta}{5} \bigg)^5, \theta = 0, 1, 2, 3, 4, 5
\]
\end{example}

$L(\theta) = k(\overline{y})f(\overline{y}; \theta)$, saa riippua aineistosta muttei parametrista!

Logaritminen uskottavuusfunktio: $l(\theta) = \logL(\theta)$, $L(\theta) > 0$

\[
\log \bigg( \prod_{i=1}^n f_{Y_i}(y_i; \theta) \bigg) = \sum_{i=1}^n \log(f_{Y_i}(y_i; \theta))
\]

$L(\hat{\theta}) \geq L(\theta)$, $\forall \theta \in \Theta$

Diskreetti: $P_{\hat{\theta}}(\overline{Y} = \overline{y}) \geq P_{\theta}(\overline{Y} = \overline{y})$, $\forall \theta \in \Theta$
